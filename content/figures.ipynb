```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array([21, 22, 12, 18]).reshape(-1, 1)
y = np.array([6,   7,  4,  5])

lm = LinearRegression().fit(x, y)
m = lm.coef_
b = lm.intercept_

yhat = m * x + b

```

```python
plt.scatter(x, y, label='actual')

plt.plot(x, yhat, c='firebrick')
plt.scatter(x, yhat, s=150, c='firebrick', marker='x', label='prediction')

plt.xlim(10, 25)
plt.ylim(2, 10)
plt.xlabel('Number of Students')
plt.ylabel('Number of Pizzas')

plt.legend()
```

```python
plt.scatter(x, y, label='actual')

plt.plot(x, yhat, c='firebrick')
plt.scatter(x, yhat, s=150, c='firebrick', marker='x', label='prediction')

plt.xlim(10, 25)
plt.ylim(2, 10)
plt.xlabel('Number of Students')
plt.ylabel('Number of Pizzas')

plt.legend()

plt.hlines(lm.predict([[16]]), 10, 16, ls='--')
plt.vlines(16, 2, lm.predict([[16]]), ls='--')
```

```python
plt.scatter(x, y)

plt.xlim(10, 25)
plt.ylim(2, 10)
plt.xlabel('Number of Students')
plt.ylabel('Number of Pizzas')

plt.vlines(16, 2, 25, ls='--')

plt.scatter(x[2:], y[2:], label='2 Closest neighbors', c='magenta')
plt.legend()
```
